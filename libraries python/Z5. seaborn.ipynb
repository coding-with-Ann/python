## count plot

```python
#It gives the no of data points are there for each category in a specific column
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(data=titanic, x='survived')
#0 means person has not survived, 1 means person has survived
```

```python
#bar plot
sns.barplot(data = titanic, x='sex', y='survived', hue='class')
```

```python
#house price dataset
from sklearn.datasets import load_diabetes
diabetes_disease = load_diabetes()
# datasets from sklearn will be in the form of numpy array, we need to convert it into dataframe
print(diabetes_disease)
```

```python
diabetes = pd.DataFrame(diabetes_disease.data, columns=diabetes_disease.feature_names)
diabetes['progression one year after'] = diabetes_disease.target
#in numpy array the progression one year after column is missing, it is stored in a array with name target
#so we created a new column and stored that data into new column called progression one year after
diabetes.head()
```

## Distribution plot(dist plot)

```python
# distplot tells in what values most of the data points lie
sns.displot(diabetes['progression one year after'])
```

## Correlation
- positive correlation
- negative correlation
- if x increases and also y increases ---> positive correlation
- if x increases and y decreases ---> negative correlation

## Heatmap
- we can plot correlation using 'heatmap'

```python
correlation = diabetes.corr()
```

```python
#constructing heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')

#If color is light then its negatively correlated, if dark color then positively correlated
#cbar is the vertical color bar on right side
#fmt means no of floating points after decimal, 1f means 1 floating point
#annot=True gives numbers inside each square
#cmap='blue' means map appears in blue color
# square=True makes each cell as a square, if false it looks like rectangle
```
