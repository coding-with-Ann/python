### correlation with heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

helps to find the inter relationship b/w features
corelation can be found only if values are either int or float 
correlation value ranges between -1 to 1
This is called pearson coeffcient correlation

```python
df.dtypes
```

```python
df.corr()
# we will get only total_bill, tip columns because they are int or float remaining are catogorical values
```

```python
sns.heatmap(df.corr())
```

## JointPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#we cant exactly call it univariate b/c we have two features in it: tip, total bill.
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pair plots are just elaborations on this, showing all variables paired with all the other variables

```python
#if the features are more than two we will be using pair plot
#again the feature need to be either int or float
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
# hue = sex indicates the plot is dependent on sex 
```

```python
df['smoker'].value_counts()
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
# kde is kernel density estimation
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
# count plot
sns.countplot(x='sex', data= df)
# if we give y = 'sex' graph will be plotted along y-axis
# count plot needs only x-axis value or y-axis value
```

```python
#barplot
sns.barplot(x= 'total_bill', y='sex', data=df)
```

```python
#bar plot
sns.barplot(x= 'sex', y='total_bill', data=df)
```

```python
df.head()
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker', y='total_bill', data=df)

#The first line in boxplot indicates 0 percentile, second line indicates 25 percentile next 50,75 and 100
# dots are outliers
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
