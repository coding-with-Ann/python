## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
#if we use this line we dont have to write plt.show()
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)

a = np.arange(40,50)
b = np.arange(50,60)
```

```python
##plt scatter

plt.scatter(x,y,c='r')
# if we dont write %matplotlib inline we have to write plt.show() here
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2-d plot')
plt.savefig('test.png') #saves image as png(also we can save image by right click on image)
```

```python
#plt plot
y = x*x
plt.plot(x,y, 'r*-', linewidth=2, markersize=12) # -- returns line in dashed format and r is red color
#r*- to print start and dashed format
#ro to print in dot format
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.title('2d plot')
```

```python
#creating subplots = creating multiple subplots in one canvas

plt.subplot(2,2,1) #2 indicates 2 rows, 2 indicates 2 columns, 1 indicates index, in canvas
plt.plot(x,y,'r--', linewidth=2)

plt.subplot(2,2,2)
plt.plot(x,y,'g*-')

plt.subplot(2,2,3)
plt.plot(x,y,'bo') #o means dotted format b is blue

plt.subplot(2,2,4)
plt.plot(x,y,'b')
```

```python
x=np.arange(1,11)
y = 3*x+5
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y,'r')
plt.title('matplot')
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin) #plots sine
plt.title('sine')

plt.subplot(2,1,2)

plt.plot(x,y_cos) #plots cos
plt.title('cos')
```

## Box Plot using Matplotlib

```python
# box plots helps to find percentiles

data = [np.random.normal(0, std, 100) for std in range(1, 4)]  #std is standard deviation, 100 is step size
# 0 is the mean (average) of the normal distribution. In this case, 
# it’s set to 0, which means that the generated random values will be centered around zero.
# The std variable stands for standard deviation. It determines how spread out the
#  data points are from the mean. we generate three datasets with different 
# standard deviations: 1, 2, and 3.
# 100: This represents the sample size. We generate 100 random values for each dataset. 
# You can adjust this number to generate more or fewer data points.
# and standard deviations of 1, 2, and 3, respectively.

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
#patch_artist = false means removes color, vert = true means boxes should be vertcal
```

## py chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.2f%%', shadow=True)

# if we sum all sizes list it will be 800, (215/800)100 is python percentage
#similarly others are calculated

#explode moves the piece a little out of chart, depends on the value we provide
# autopct indicates data type here float

plt.axis('equal')
plt.show()
```
